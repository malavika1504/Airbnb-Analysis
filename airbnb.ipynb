```python
#host
host= []
for i in coll.find({},{"_id":1, "host":1}):
    host.append(i)
```

```python
df_host= pd.DataFrame(host)
df_host
```

```python
hostcol= {'_id':[],'host_id':[], 'host_url':[], 'host_name':[], 'host_location':[],"host_response_time":[], 'host_thumbnail_url':[], 'host_picture_url':[], 'host_neighbourhood':[], 'host_response_rate':[], 'host_is_superhost':[], 'host_has_profile_pic':[], 'host_identity_verified':[], 'host_listings_count':[], 'host_total_listings_count':[], 'host_verifications':[]}

for i in df_host["_id"]:
    hostcol["_id"].append(i)
for i in df_host["host"]:
    hostcol["host_id"].append(i["host_id"])
    hostcol["host_url"].append(i["host_url"])
    hostcol["host_name"].append(i["host_name"])
    hostcol["host_location"].append(i.get("host_location", ""))
    hostcol["host_response_time"].append(i.get("host_response_time"))
    hostcol["host_thumbnail_url"].append(i["host_thumbnail_url"])
    hostcol["host_picture_url"].append(i["host_picture_url"])
    hostcol["host_neighbourhood"].append(i.get("host_neighbourhood"))
    hostcol["host_response_rate"].append(i.get("host_response_rate"))
    hostcol["host_is_superhost"].append(i["host_is_superhost"])
    hostcol["host_has_profile_pic"].append(i["host_has_profile_pic"])
    hostcol["host_identity_verified"].append(i["host_identity_verified"])
    hostcol["host_listings_count"].append(i["host_listings_count"])
    hostcol["host_total_listings_count"].append(i["host_total_listings_count"])
    hostcol["host_verifications"].append(i.get("host_verifications", []))
```

```python
df_host1= pd.DataFrame(hostcol)
df_host1.head(6)
```

```python
df_host1.isnull().sum()
```

```python
df_host1["host_response_time"].fillna("Unspecified",inplace= True)
df_host1["host_response_rate"].fillna("Unspecified",inplace= True)
df_host1["host_neighbourhood"].fillna("Unspecified",inplace= True)
```

```python
df_host1["host_is_superhost"]= df_host1["host_is_superhost"].map({False: "No", True: "Yes"})
df_host1["host_has_profile_pic"]= df_host1["host_has_profile_pic"].map({False: "No", True: "Yes"})
df_host1["host_identity_verified"]= df_host1["host_identity_verified"].map({False: "No", True: "Yes"})
```

```python
df_host1.isnull().sum()
```

```python
#ADDRESS
address= []
for i in coll.find({}, {"_id":1, "address":1}):
    address.append(i)
```

```python
df_address=pd.DataFrame(address)
df_address.head()
```

```python
addresscol= {'_id':[], 'street':[], 'suburb':[], 'government_area':[], 'market':[], 'country':[],
                   'country_code':[], 'location_type':[], "longitude":[], "latitude":[],
                    "is_location_exact":[]}

for i in df_address["_id"]:
    addresscol["_id"].append(i)

for i in df_address["address"]:
    addresscol["street"].append(i["street"])
    addresscol["suburb"].append(i.get("suburb",""))
    addresscol["government_area"].append(i["government_area"])
    addresscol["market"].append(i.get("market",""))
    addresscol["country"].append(i["country"])
    addresscol["country_code"].append(i["country_code"])
    addresscol["location_type"].append(i["location"]["type"])
    addresscol["longitude"].append(i["location"]["coordinates"][0])
    addresscol["latitude"].append(i["location"]["coordinates"][1])
    addresscol["is_location_exact"].append(i["location"]["is_location_exact"])
```

```python
df_address1=pd.DataFrame(addresscol)
df_address1.head()
```

```python
df_address1.isnull().sum()
```
